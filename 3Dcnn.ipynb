```python
import os
import cv2
import numpy as np
import torch
import pandas as pd
import concurrent.futures
from tqdm import tqdm

#Image processing
root_dir = '/content/drive/MyDrive/Dataset/images2'

def load_image(frame_number, root_dir=root_dir, extension='jpg'):
    # Construct the file path
    image_path = os.path.join(root_dir, f"{frame_number}.{extension}")
    # Load the image
    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    if image is None:
        raise FileNotFoundError(f"Image not found: {image_path}")
    return image

def preprocess_image(image, size=(140, 240)):
    # Convert to grayscale
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Check the current size and resize if necessary
    if image.shape[:2] != size:
        print(f"Resizing image from {image.shape[:2]} to {size}")
        image = cv2.resize(image, size)
    # Normalize the image
    image = image / 255.0
    return image

# Hardcode the clip frames
clip_frames = {
    1: (1, 207),
    2: (208, 337),
    3: (338, 400),
    4: (401, 508),
    5: (509, 668),
    6: (669, 746),
    7: (747, 1198),
    8: (1199, 1313),
    9: (1314, 1422),
    10: (1423, 1552),
    11: (1553, 1678),
    12: (1679, 1931),
    13: (1932, 2049),
    14: (2050, 2144),
    15: (2145, 2527),
    16: (2528, 2660),
    17: (2661, 2739),
    18: (2740, 2805),
    19: (2806, 3227),
    20: (3228, 3546),
    21: (3547, 3784)
}
#Relating to temporal aspect
# Create a dictionary to store clips with the correct number of frames
clips = {}
for clip_num, (start_frame, end_frame) in clip_frames.items():
    clips[clip_num] = [(frame, clip_num) for frame in range(start_frame, end_frame + 1)]

```

```python
def preprocess_frame(frame_label_tuple):
    frame, label = frame_label_tuple
    image = preprocess_image(load_image(frame))
    return (frame, label, image)

# Preprocess clips in parallel with progress bar
for clip_num, frames in clips.items():
    with concurrent.futures.ThreadPoolExecutor() as executor: #this is what parallelises
        clips[clip_num] = list(tqdm(executor.map(preprocess_frame, frames), total=len(frames), desc=f'Processing clip {clip_num}'))#function to execute the preprocessing adding a  progress bar

def pad_clip(clip, length, image_size=(140, 240)):
    padded_clip = np.zeros((length, 1, image_size[0], image_size[1]))  # Adding channel dimension
    for i, (_, _, image) in enumerate(clip):
        if image.shape == (image_size[0], image_size[1]):  # Ensure the image has the correct shape
            padded_clip[i, 0] = image
        else:
            print(f"Skipping image with incorrect shape: {image.shape}")
    return padded_clip
#Loading and padding for consistent dimensions
# Pad all clips to their specified lengths
max_length = max(end_frame - start_frame + 1 for start_frame, end_frame in clip_frames.values())
padded_clips = {clip_num: pad_clip(frames, max_length) for clip_num, frames in clips.items()}
```

```python
import os
import cv2
import numpy as np
import torch
import pandas as pd
import concurrent.futures
from tqdm import tqdm
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split

#Access the saved tensors
save_dir = '/content/drive/MyDrive/Dataset'

# Load the tensors from Google Drive
clip_tensors = torch.load(f'{save_dir}/clip_tensors.pt')
label_tensors = torch.load(f'{save_dir}/label_tensors.pt')

class TennisDataset(Dataset):
    def __init__(self, clips, labels):
        self.clips = clips
        self.labels = labels

    def __len__(self):
        return len(self.clips)

    def __getitem__(self, idx):
        clip = self.clips[idx]
        label = self.labels[idx]
        return clip, label

# Create dataset
dataset = TennisDataset(clip_tensors, label_tensors)

# Split dataset into train and test sets
train_size = int(0.5 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])

# Create data loaders
batch_size = 1  # Reduce batch size to lower memory usage
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class Optimized3DCNN(nn.Module):
    def __init__(self, input_shape):
        super(Optimized3DCNN, self).__init__()
        self.conv1 = nn.Conv3d(1, 8, kernel_size=(3, 3, 3), stride=(1, 1, 1), padding=(1, 1, 1))
        self.pool = nn.MaxPool3d(kernel_size=(2, 2, 2), stride=(2, 2, 2))
        self.conv2 = nn.Conv3d(8, 16, kernel_size=(3, 3, 3), stride=(1, 1, 1), padding=(1, 1, 1))

        # Calculate the size of the tensor after conv and pool layers
        self._to_linear = self._get_conv_output(input_shape)
        self.fc1 = nn.Linear(self._to_linear, 128)
        self.fc2 = nn.Linear(128, len(set(label_tensors.numpy())))  # Number of classes

    def _get_conv_output(self, shape):
        # Create a dummy tensor with the shape
        bs = 1  # batch size
        input = torch.rand(bs, *shape)
        output_feat = self.conv1(input)
        output_feat = self.pool(output_feat)
        output_feat = self.conv2(output_feat)
        output_feat = self.pool(output_feat)
        n_size = output_feat.view(1, -1).size(1)
        return n_size

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        x = F.relu(self.conv2(x))
        x = self.pool(x)
        x = x.view(-1, self._to_linear)  # Flatten
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Get the shape from the loaded tensor
input_shape = (1, 452, 140, 240)  # (channels, depth, height, width)
model = Optimized3DCNN(input_shape)

# Print the calculated dimension for the fully connected layer
print(f"The size of the tensor after convolution and pooling layers is: {model._to_linear}")
```

```python
# Initialize model, loss function, and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 10  # Adjust as needed
accumulation_steps = 4  # Number of steps to accumulate gradients

for epoch in range(num_epochs):
    epoch_loss = 0.0
    optimizer.zero_grad()  # Initialize gradients

    for i, (clips, labels) in enumerate(train_loader):
        clips = clips.permute(0, 2, 1, 3, 4)  # Change to (batch_size, channels, depth, height, width)

        # Forward pass
        outputs = model(clips)
        loss = criterion(outputs, labels)

        # Backward pass with gradient accumulation
        loss = loss / accumulation_steps
        loss.backward()

        if (i + 1) % accumulation_steps == 0:
            optimizer.step()
            optimizer.zero_grad()

        epoch_loss += loss.item() * accumulation_steps

        # # Explicitly free memory
        # del clips, labels, outputs, loss
        # torch.cuda.empty_cache()  # If using GPU, otherwise not needed

    avg_epoch_loss = epoch_loss / len(train_loader)
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {avg_epoch_loss:.4f}')
```
